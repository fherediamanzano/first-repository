#### 3)Medidas resumen del la base de datos

#### Descripción general de la BBDD

```python
serie_olimpica.describe()
```

#### Sumo por país la cantidad de medallas ganadas en el período analizado

```python
#Muestra total medallas pero no ordenado
serie_olimpica.groupby('Country')['Medal'].count() 
```

#### Si ordeno los paises en decreciente segun la cantidad de medallas ganadas para el período analizado

```python
#Agrego para ordenar. Por defecto es ascendente
serie_olimpica.groupby('Country')['Medal'].count().sort_values(ascending=False)
```

#### Si solo quiero mostrar el 'top 3' paises con más medallas

```python
#Muestro el top 3
serie_olimpica.groupby('Country')['Medal'].count().nlargest(3)
```

#### Otra forma de mostrar en orden decreciente la cantidad de medallas ganadas por paises

```python
serie_olimpica['Country'].value_counts()
```

#### 4)Tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento

#### Dada la serie olímpica

```python
serie_olimpica
```

#### Se filtra por 'Event_gender' = 'M' y se agrupa por año el total de medallas

```python
serie_olimpica[serie_olimpica['Event_gender']=='M'].groupby('Year')['Medal'].count()
```

#### Si quisiera saber el total de medallas de oro ganadas por hombre a lo largo de los años, sería

```python
serie_olimpica[(serie_olimpica['Event_gender']=='M')&(serie_olimpica['Medal']=='Gold')].groupby('Year')['Medal'].count()
```
