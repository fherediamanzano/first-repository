#### Clase 6 introducci√≥n a Pandas

```python
import pandas as pd
```

```python
df_lluvias = pd.read_csv("pune_1965_to_2002.csv")
df_lluvias
```

```python
indice = list(df_lluvias.Year)
indice
```

```python
columnas = df_lluvias.columns[1:]
columnas
```

```python
valores=df_lluvias.values[:,1:]
```

```python
valores
```

```python
df_lluvias_final = pd.DataFrame(valores,index=indice,columns=columnas)
df_lluvias_final
```

```python
df_lluvias_final.sum()
```

```python
df_lluvias_final.mean(axis='columns')
```

```python
df_lluvias_final.describe().round(1)
```

```python
df_lluvias_final.T.describe().round(1)
```
