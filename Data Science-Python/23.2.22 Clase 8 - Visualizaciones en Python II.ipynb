```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

##### Ejemplos de subgr√°ficos
Primero traemos el data set

```python
df_lluvias = pd.read_csv('pune_1965_to_2002.csv')
df_lluvias.head()
```

```python
df_lluvias.index = df_lluvias['Year']
df_lluvias = df_lluvias.drop(['Year'], axis='columns')
df_lluvias
```

```python
#Metemos un grafico que compare 3 series
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(12, 5), sharex=True, sharey=True)
#eje x ponemos los a√±os
x = df_lluvias.index
#eje y ponemos las lluvias para cada mes a comparar
ax[0].plot(df_lluvias.index, df_lluvias['Jan'], label='Precipitaciones de enero')
ax[1].plot(df_lluvias.index, df_lluvias['Feb'], label='Precipitaciones de febrero', color='C1')
ax[2].plot(df_lluvias.index, df_lluvias['Mar'], label='Precipitaciones de marzo', color='C2')
#a√±adimos texto
ax[0].set_title('Precipitaciones de los primeros tres meses del a√±o') 
ax[2].set_xlabel('A√±o')  
ax[1].set_ylabel('Precipitaci√≥n (mm.)')
#agregamos las leyendas
ax[0].legend()  
ax[1].legend()
ax[2].legend() 
```

#### Trae datasets de prueba

```python
df_ejercicio = sns.load_dataset('exercise')
df_ejercicio = df_ejercicio.drop('Unnamed: 0', axis='columns')
df_ejercicio.head()
```

#### Traigo los que entrenaron s√≥lo 30 minutos

```python
df_30_min = df_ejercicio[df_ejercicio['time'] == '30 min']
df_30_min.head()
```

#### Graficamos las distribuciones

```python
plt.figure()
ax = sns.displot(data=df_30_min, x='pulse', kind='kde', hue='kind', fill=True)
ax.set(xlabel='Frecuencia Card√≠aca', ylabel='Densidad', title='Distribuci√≥n  de las pulsaciones')
```

```python
plt.figure()
ax = sns.displot(data=df_30_min, x='pulse', kind='hist', hue='kind', fill=True)
ax.set(xlabel='Frecuencia Card√≠aca', ylabel='Densidad', title='Distribuci√≥n  de las pulsaciones')
```

```python
plt.figure()
ax = sns.displot(data=df_30_min, x='pulse', kind='kde', hue='diet', fill=True)
ax.set(xlabel='Frecuencia Card√≠aca', ylabel='Densidad', title='Distribuci√≥n  de las pulsaciones')
```

#### Visualizar los datos en base a dos variables distintas
sns.catplot <br>
kind='violin'<br>
Visualizamos pulsaciones en base a actividad y tipo de dieta

```python
ax = sns.catplot(data=df_ejercicio, kind='violin', x='time', y='pulse', hue='diet', split=True)
ax.set(xlabel='Duraci√≥n de ejercicio', ylabel='Frecuencia card√≠aca', title='Categorizaci√≥n de la distribuci√≥n de pulsaciones')
```

#### Para visualizar en base a la actividad le cambiamos el valor de x

#### 2)Histograma de salarios

#### Dada la columna de los salarios del personal

```python
df_HR['Salary']
```

#### Se grafica el histograma

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(df_HR['Salary'].values.flatten(), bins=10)#Grafico de histograma
ax.set_title('Histograma de salarios')
ax.set_xlabel('Intervalos de salarios [$]')
ax.set_ylabel('Frecuencia absoluta')
```

#### 3)Grafico viol√≠n con los salarios discriminados por g√©nero

```python
plt.figure()
ax = sns.catplot(data=df_HR, kind='violin', y='Salary', hue='GenderID', split=True)
ax.set(xlabel='Duraci√≥n de ejercicio', ylabel='Frecuencia card√≠aca', title='Categorizaci√≥n de la distribuci√≥n de pulsaciones')
```

#### 4)Contrataciones por fecha

```python
df_HR['DateofHire']
```

#### Convierto a formato fecha

```python
df_HR['DateofHire'] = pd.to_datetime(df_HR['DateofHire'])
df_HR['DateofHire']
```

```python
df_HR['Year'] = df_HR['DateofHire'].dt.year 
df_HR['Year']
```
