#### Parece simple al principio, luego se complica

#### Siempre incluir informaci√≥n de lo que se muestra; nombre de ejes, titulo, leyenda

#### Importo los datasets

```python
df_lluvias = pd.read_csv('pune_1965_to_2002.csv')
df_lluvias.head()
```

#### Ponemos a Year como indice, y eliminamos el anterior

```python
df_lluvias.index = df_lluvias['Year']
df_lluvias
```

```python
df_lluvias = df_lluvias.drop(['Year'], axis='columns')
df_lluvias
```

#### Graficamos

```python
x = df_lluvias.index
x
```

```python

y = df_lluvias.sum(axis='columns')
y
```

#### Definimos los objetos fig y ax

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x,y,label='Precipitaciones acumuladas')
```

#### A√±adimos etiqetas a ejes

```python
ax.set_xlabel('A√±o')
ax.set_ylabel('Precipitaci√≥n acumulada (mm.)')
ax.set_title('Precipitaciones acumuladas seg√∫n el a√±o')
ax.legend(loc='lower right')#cambio localizaci√≥n de la etiqueta
fig #Llamo al gr√°fico
```

##### Recortar bordes vacios del gr√°fico

```python
ax.set_xlim(df_lluvias.index[0],df_lluvias.index[-1])
fig
```

##### Guardar gr√°ficos

```python
#Puede ser formato jpeg, jpg, png, pdf, svg)
fig.savefig("precipitaciones_del_1965_al_2002.pdf")
```

####  Otro grafico
Ahora vemos la corelaci√≥n entre Agosto y Septiembre

```python
fig, ax = plt.subplots()  
mapeo_colores = ax.scatter(df_lluvias['Aug'], df_lluvias['Sep'], c=df_lluvias.index)
fig.colorbar(mapeo_colores)#Grafico puntos
ax.set_title('Precipitaciones Agosto-Septiembre')
ax.set_xlabel('Precipitaciones en Agosto (mm.)')
ax.set_ylabel('Precipitaciones en Septiembre (mm.)')
```

#### Grafico de barras

```python
precipitaciones_acumuladas = df_lluvias.sum()
precipitaciones_acumuladas
```

```python
fig, ax = plt.subplots(figsize=(8,4))
precipitaciones_acumuladas = df_lluvias.sum()
ax.bar(df_lluvias.columns, precipitaciones_acumuladas)#Grafico de barras
ax.set_title('Precipitaciones acumuladas desde 1965 a 2002, seg√∫n el mes')
ax.set_ylabel('Precipitaci√≥n total (mm.)')
ax.set_xlabel('Mes')
```

#### Multiples elementos

```python
fig, ax = plt.subplots(figsize=(12, 3))  
ax.plot(df_lluvias.index, df_lluvias['Jan'], label='Precipitaciones de enero')
ax.plot(df_lluvias.index, df_lluvias['Feb'], label='Precipitaciones de febrero', color='C1')
```

#### Calculamos el m√°ximo de cada serie

```python
maximo_enero = df_lluvias['Jan'].max()
maximo_enero
```

```python
maximo_febrero =  df_lluvias['Feb'].max()
maximo_febrero
```

#### Dibujamos el m√°ximo de cada serie

```python
fig, ax = plt.subplots(figsize=(12, 3))  
ax.plot(df_lluvias.index, df_lluvias['Jan'], label='Precipitaciones de enero')
ax.plot(df_lluvias.index, df_lluvias['Feb'], label='Precipitaciones de febrero', color='C1')
#Agregamos lineas de m√°ximos
ax.axhline(maximo_enero, linestyle='--', alpha=0.5,    linewidth=3, label='M√°xima de enero')
ax.axhline(maximo_febrero, color='C1', linestyle=':', alpha=0.5, linewidth=3, label='M√°xima de febrero')
```

```python
fig, ax = plt.subplots(figsize=(12, 3))  
ax.plot(df_lluvias.index, df_lluvias['Jan'], label='Precipitaciones de enero')
ax.plot(df_lluvias.index, df_lluvias['Feb'], label='Precipitaciones de febrero', color='C1')
#Agregamos lineas de m√°ximos
ax.axhline(maximo_enero, linestyle='--', alpha=0.5,    linewidth=3, label='M√°xima de enero')
ax.axhline(maximo_febrero, color='C1', linestyle=':', alpha=0.5, linewidth=3, label='M√°xima de febrero')
#Agregamos labels
ax.set_xlabel('A√±o')  
ax.set_ylabel('Precipitaci√≥n (mm.)') 
ax.set_title('Precipitaciones de enero y febrero') 	
#Eliminamos espacios vacios
ax.set_xlim(df_lluvias.index[0], df_lluvias.index[-1])
#Mostramos los nombres de las l√≠neas dibujadas
ax.legend() 
```
